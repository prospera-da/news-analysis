**Define Functions as reusable codes**

```python
# Function to aggregate the dataset
def aggregate_data(df):
    """Compute aggregate metrics from the Twitter dataset."""
    if df is not None:
        # Aggregate metrics
        total_tweets = len(df)
        total_retweets = df["retweetCount"].sum()
        total_replies = df["replyCount"].sum()
        total_bookmarks = df["bookmarkCount"].sum()

        # Find the tweet with the most retweets and most replies
        most_retweeted = df.sort_values(by="retweetCount", ascending=False).iloc[0]
        most_replied = df.sort_values(by="replyCount", ascending=False).iloc[0]

        print("Aggregated Twitter Metrics:")
        print(f"Total Tweets: {total_tweets}")
        print(f"Total Retweets: {total_retweets}")
        print(f"Total Replies: {total_replies}")
        print(f"Total Bookmarks: {total_bookmarks}")

        print("\nTop Tweets:")
        print(
            f"Most Retweeted: {most_retweeted['fullText']} (Retweets: {most_retweeted['retweetCount']})"
        )
        print(
            f"Most Replied: {most_replied['fullText']} (Replies: {most_replied['replyCount']})"
        )
    else:
        print("No data to aggregate.")
```

**Main Execution**

```python
file_path = "dataset_twitter-scraper-lite.xlsx"

# Read and Load excel file
dataset = pd.read_excel(file_path)
dataset.head(5)
```

```python
aggregate_data(dataset)
```
